# Week 1: Data Acquisition, Cleaning, and Feature Engineering

In this notebook, we will:
1. Load the raw Hotel Booking Demand dataset.
2. Handle missing values in columns like `agent` and `company`.
3. Treat extreme outliers in `adr` (Average Daily Rate).
4. Engineer new features, such as `total_duration` (weekend + weekday nights).

```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os

# Ensure plot outputs appear in the notebook
%matplotlib inline
```

## 1. Load Data

```
file_path = '../data/hotel_bookings.csv'
if os.path.exists(file_path):
    df = pd.read_csv(file_path)
    print('Dataset shape:', df.shape)
    display(df.head())
else:
    print('Dataset not found. Please run setup_data.py first.')
```

## 2. Handle Missing Values

```
# Check missing values
missing_vals = df.isnull().sum()
print(missing_vals[missing_vals > 0])

# 'company' has very high missing values, we might drop it or fill with 0 (meaning no company)
df['company'] = df['company'].fillna(0)

# 'agent' can also be filled with 0
df['agent'] = df['agent'].fillna(0)

# 'country' filled with 'Unknown'
df['country'] = df['country'].fillna('Unknown')

# 'children' missing values are small, fill with 0
df['children'] = df['children'].fillna(0)

print('\nAfter filling:\n', df[['company', 'agent', 'country', 'children']].isnull().sum())
```

## 3. Treat Outliers in Average Daily Rate (ADR)

```
plt.figure(figsize=(10,4))
sns.boxplot(x=df['adr'])
plt.title('ADR Distribution (Before Cleaning)')
plt.show()

# Remove negative ADR and extreme outliers (e.g. > 5000)
df = df[(df['adr'] >= 0) & (df['adr'] < 5000)]

plt.figure(figsize=(10,4))
sns.boxplot(x=df['adr'])
plt.title('ADR Distribution (After Cleaning)')
plt.show()
```

## 4. Feature Engineering

```
# Calculate total duration of stay
df['total_duration'] = df['stays_in_weekend_nights'] + df['stays_in_week_nights']

# Remove records with zero duration
df = df[df['total_duration'] > 0]

print('Final dataset shape after cleaning:', df.shape)
df.head()
```

## 5. Save Cleaned Data

```
df.to_csv('../data/cleaned_hotel_bookings.csv', index=False)
print('Cleaned data saved to data/cleaned_hotel_bookings.csv')
```
