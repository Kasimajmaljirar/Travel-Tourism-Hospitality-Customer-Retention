# Week 2: Exploratory Data Analysis (EDA) and Statistical Testing

In this notebook, we will:
1. Perform Univariate and Bivariate analysis.
2. Analyze the relationship between Average Daily Rate (ADR) and cancellation rates.
3. Map correlation matrices to identify drivers of cancellations.
4. Visualize the booking curve (lead time vs. booking volume).

```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline

# Set seaborn aesthetic style
sns.set_style('whitegrid')
```

## 1. Load Cleaned Data

```
df = pd.read_csv('../data/cleaned_hotel_bookings.csv')
df.head()
```

## 2. Cancellation Overview

```
plt.figure(figsize=(6,4))
sns.countplot(x='is_canceled', data=df, palette='Set2')
plt.title('Booking Status (0 = Not Canceled, 1 = Canceled)')
plt.show()

cancel_rate = df['is_canceled'].mean() * 100
print(f'Overall Cancellation Rate: {cancel_rate:.2f}%')
```

## 3. ADR vs Cancellation Rate (Bivariate Analysis)
We explore whether higher daily rates correlate with higher cancellation rates.

```
plt.figure(figsize=(8,5))
sns.boxplot(x='is_canceled', y='adr', data=df[df['adr'] < 500], palette='Set1')
plt.title('ADR Distribution by Cancellation Status')
plt.show()
```

## 4. Booking Curve (Lead Time vs Booking Volume)

```
plt.figure(figsize=(10,6))
sns.histplot(df['lead_time'], bins=50, kde=True, color='skyblue')
plt.title('Distribution of Lead Time (Booking Curve)')
plt.xlabel('Lead Time (Days)')
plt.ylabel('Number of Bookings')
plt.show()
```

## 5. Correlation Matrix
Let's isolate numeric features to find strong indicators of cancellation.

```
numeric_cols = df.select_dtypes(include=[np.number])
corr = numeric_cols.corr()

plt.figure(figsize=(12, 10))
sns.heatmap(corr, annot=False, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Matrix of Numeric Features')
plt.show()

print('Top numerical correlations with cancellation:')
print(corr['is_canceled'].sort_values(ascending=False).head(10))
```
